````python
def get_enhanced_recommendations(movie_title, num_recommendations=5):
    """
    Enhanced recommendation function using content-based filtering + Gemini AI
    """
    try:
        # Get index of the movie
        index = new_df[new_df['title'] == movie_title].index[0]
        
        # Calculate similarity scores
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
        
        # Get top recommended movies
        recommended_movies = []
        for i in distances[1:num_recommendations+1]:
            movie_info = {
                'title': new_df.iloc[i[0]].title,
                'similarity_score': round(i[1], 3),
                'tags': new_df.iloc[i[0]].tags[:200] + "..." if len(new_df.iloc[i[0]].tags) > 200 else new_df.iloc[i[0]].tags
            }
            recommended_movies.append(movie_info)
        
        # Create a detailed prompt for Gemini
        movies_list = "\n".join([f"• {movie['title']} (Similarity: {movie['similarity_score']})" for movie in recommended_movies])
        original_movie_tags = new_df.iloc[index].tags[:300] + "..." if len(new_df.iloc[index].tags) > 300 else new_df.iloc[index].tags
        
        prompt = f"""
        I have a movie recommendation system that found these similar movies to "{movie_title}":
        
        Original Movie Tags: {original_movie_tags}
        
        Recommended Movies:
        {movies_list}
        
        Please provide:
        1. A brief explanation of why these movies are similar to "{movie_title}"
        2. Which of these recommendations would be best for different moods (action, drama, comedy, etc.)
        3. Rate each recommendation from 1-10 for similarity and explain why
        
        Make it engaging and helpful for movie lovers!
        """
        
        # Get Gemini's analysis
        response = model.generate_content(prompt)
        
        return {
            'original_movie': movie_title,
            'recommendations': recommended_movies,
            'ai_analysis': response.text
        }
        
    except IndexError:
        return {
            'error': f"Movie '{movie_title}' not found in database. Try another title."
        }
    except Exception as e:
        return {
            'error': f"An error occurred: {str(e)}"
        }
````

````python
def search_movies_by_description(user_query, top_k=10):
    """
    Search movies using natural language descriptions via Gemini AI
    """
    # Get all movie titles and their tags
    movies_data = []
    for idx, row in new_df.iterrows():
        movies_data.append({
            'title': row['title'],
            'tags': row['tags'][:150] + "..." if len(row['tags']) > 150 else row['tags']
        })
    
    # Create sample of movies for Gemini to analyze (to avoid token limits)
    sample_movies = movies_data[:200] if len(movies_data) > 200 else movies_data
    movies_text = "\n".join([f"Title: {movie['title']}\nTags: {movie['tags']}\n" for movie in sample_movies])
    
    prompt = f"""
    Based on this user query: "{user_query}"
    
    From the following movie database, recommend the top {top_k} movies that best match the user's request:
    
    {movies_text}
    
    Please analyze the user's preferences and provide:
    1. Top {top_k} movie recommendations with titles exactly as they appear in the database
    2. Brief explanation for each recommendation
    3. Overall analysis of what the user is looking for
    
    Format your response as:
    RECOMMENDATIONS:
    1. [Movie Title] - [Brief explanation]
    2. [Movie Title] - [Brief explanation]
    etc.
    
    ANALYSIS:
    [Your analysis of user preferences]
    """
    
    try:
        response = model.generate_content(prompt)
        return {
            'query': user_query,
            'ai_response': response.text,
            'total_movies_analyzed': len(sample_movies)
        }
    except Exception as e:
        return {
            'error': f"Search failed: {str(e)}"
        }
````

````python
# Complete Data Processing and Model Training Pipeline
print("🚀 COMPLETE MOVIE RECOMMENDATION SYSTEM PIPELINE")
print("="*60)

import numpy as np
import pandas as pd
import ast
import pickle
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import google.generativeai as genai

# Step 1: Load Data
print("📊 Step 1: Loading movie data...")
try:
    movies = pd.read_csv('data/tmdb_5000_movies.csv')
    credits = pd.read_csv('data/tmdb_5000_credits.csv')
    print(f"✅ Loaded {len(movies)} movies and {len(credits)} credit records")
except FileNotFoundError as e:
    print(f"❌ Error loading data: {e}")
    print("Please ensure the CSV files are in the 'data' folder")
    raise

# Step 2: Merge and preprocess
print("\n🔧 Step 2: Merging and preprocessing data...")
movies = movies.merge(credits, on='title')
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.dropna(inplace=True)
print(f"✅ After cleaning: {len(movies)} movies")

# Step 3: Feature extraction functions
print("\n⚙️ Step 3: Setting up feature extraction...")

def convert(text):
    L = []
    try:
        for i in ast.literal_eval(text):
            L.append(i['name']) 
    except:
        pass
    return L

def convert_cast(text):
    L = []
    counter = 0
    try:
        for i in ast.literal_eval(text):
            if counter < 3:
                L.append(i['name'])
            counter+=1
    except:
        pass
    return L

def fetch_director(text):
    L = []
    try:
        for i in ast.literal_eval(text):
            if i['job'] == 'Director':
                L.append(i['name'])
                break
    except:
        pass
    return L

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

# Step 4: Apply transformations
print("🔄 Step 4: Applying transformations...")
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(fetch_director)
movies['overview'] = movies['overview'].apply(lambda x: x.split() if pd.notna(x) else [])

# Remove spaces
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)

# Combine all features
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Create final dataframe
new_df = movies[['movie_id','title','tags']].copy()
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).lower())

print(f"✅ Feature extraction complete for {len(new_df)} movies")

# Step 5: Vectorization and similarity calculation
print("\n🧮 Step 5: Creating similarity matrix...")
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
similarity = cosine_similarity(vector)
print(f"✅ Similarity matrix created: {similarity.shape}")

# Step 6: Configure Gemini API
print("\n🤖 Step 6: Configuring Gemini API...")
GEMINI_API_KEY = "AIzaSyDJ3TY3gZP2g8yiV40sc-3a1gevGuHUXIU"
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('gemini-1.5-flash')
print("✅ Gemini API configured")

print("\n🎉 PIPELINE COMPLETE!")
print(f"📊 Total movies processed: {len(new_df)}")
print(f"🎬 Sample movies: {', '.join(new_df['title'].head(3).tolist())}")
print(f"🧮 Feature vector size: {vector.shape[1]}")
print(f"💾 Ready to save model and test recommendations!")
````
